```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(' https://raw.githubusercontent.com/PGaea/ButardoPrincessGaea_BSIS2/main/SALARY_DATA.csv')
```

```python
print(df)
```

```python
salary = df['Salary'].to_numpy()
```

```python
mean_salary = np.mean(salary)
```

```python
print(mean_salary)
```

```python
median_salary = np.median(salary)
```

```python
print(median_salary)
```

```python
std_salary = np.std(salary)
```

```python
print(std_salary)
```

```python
total = np.sum(salary)
```
